```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
# Total political tweets by users

userAggs = {
    'tweetsDay': {
        'Tweets in Stream': 'sum'
    },
    'userID': {
        'numberDays': 'count'
    }
}

tweetsOverall = df.groupby('userID').agg(userAggs)
tweetsOverall.columns = ['numberTweets', 'numberDays']


#tweetsOverall.reset_index(drop=False, inplace=True)
#tweetsOverall.sort('numberTweets', ascending=False, inplace=True)
```

```python
#tweetsOverall.reset_index(drop=False, inplace=True)
tweetsOverall.head()
```

```python
mostRecentTweets = df.sort(['userID', 'timestamp']).drop_duplicates(subset=["userID"], take_last=False)
```

```python
#mostRecentTweets.reset_index(drop=False, inplace=True)
mostRecentTweets.head()
```

```python
tidy = pd.concat([tweetsOverall, mostRecentTweets], axis='userID', join='inner')
tidy.head()
```

```python
testAggs = {
    'tweetsDay': 'sum', 
    'tweetsAll': 'max',
    'userID': 'count',
    'followers': 'min',
    'following': 'min',
    'accountDurationDays': 'max'
}

test = df.groupby('userID').agg(testAggs)
candSplit = df.groupby(['userID', 'candidate']).agg(testAggs)
#test.reset_index(drop=False, inplace=True)
test.head()
```

```python
print test.shape
print candSplit.shape
candSplit.head()
```

```python
test.columns = ['streamTweets', 'accountDuration', 'numDays', 'allTweets', 'followers', 'following']
candSplit.columns = ['streamTweets', 'accountDuration', 'numDays', 'allTweets', 'followers', 'following']
test.reset_index(inplace=True, drop=False)
candSplit.reset_index(inplace=True, drop=False)
candSplit.head()
```

```python
test.to_csv('update/agg-users.csv', index=False)
candSplit.to_csv('update/agg-users-by-candidate.csv', index=False)
```

We don't care which stream the bot is in for the API calls, because the ID request will return either way.

```python
filtered = test[(test['numDays'] > 1) & 
     (test['streamTweets'] > 1) & 
             ((test['following'] < 15) | 
             (test['followers'] < 15) | 
             (test['accountDuration'] < 32) |
             (test['streamTweets'] > 1000))]
print filtered.shape
```

```python
filtered.sort('streamTweets', ascending=False, inplace=True)
filtered.head()
```

```python
# The candidates
filtered.to_csv('update/filtered-users.csv', index=False)
```
